```python
import pandas as pd
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

```python
X_train = pd.read_csv('X_train_scaled.csv')
X_test = pd.read_csv('X_test_scaled.csv')
y_train = pd.read_csv('y_train.csv')
y_test = pd.read_csv('y_test.csv')
X_oot = pd.read_csv('X_oot_scaled.csv')
y_oot = pd.read_csv('y_oot.csv')
```

```python
# create model
model = Sequential()
model.add(Dense(5, input_dim=20, init='uniform', activation='relu'))
model.add(Dense(4, init='uniform', activation='relu'))
model.add(Dense(1, init='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X_train, y_train, nb_epoch=20, batch_size=10)
```

```python
# calculate predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
y_oot_pred = model.predict(X_oot)
```

```python
###y_pred
```

```python
a_train=[]
prob_train = model.predict_proba(X_train)
for probabilty in prob_train:
    a_train.append(probabilty[0])
```

```python
a_test=[]
prob_test = model.predict_proba(X_test)
for probabilty in prob_test:
    a_test.append(probabilty[0])
```

```python
a_oot=[]
prob_oot = model.predict_proba(X_oot)
for probabilty in prob_oot:
    a_oot.append(probabilty[0])
```

```python
newDF_train = X_train
newDF_train['prob'] = a_train
newDF_train['org'] = y_train
newDF_train.sort_values(by=['prob'],ascending = False, inplace = True)

newDF_test = X_test
newDF_test['prob'] = a_test
newDF_test['org'] = y_test
newDF_test.sort_values(by=['prob'],ascending = False, inplace = True)

newDF_oot = X_oot
newDF_oot['prob'] = a_oot
newDF_oot['org'] = y_oot
newDF_oot.sort_values(by=['prob'],ascending = False, inplace = True)
```

```python
y_train_pred = [round(x[0]) for x in y_train_pred]
y_test_pred = [round(x[0]) for x in y_test_pred]
y_oot_pred = [round(x[0]) for x in y_oot_pred]
```
