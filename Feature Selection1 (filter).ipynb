```python
#tdata.drop(['Recnum'], axis =1, inplace =True)
tdata['Random'] = np.random.ranf(len(tdata))
tdata.shape
```

```python
goods = tdata[tdata['Fraud']==0]
goods.shape
```

```python
bads = tdata[tdata['Fraud'] == 1]
bads.shape
```

```python
numgoods = len(goods)
numbads = len(bads)
```

```python
tdata.isnull().sum()
```

```python
tdata = tdata.drop('Merch description',1)
```

```python
tdata = tdata.drop(['Cardnum','Date','Merchnum','Merch state','Merch zip','Transtype'],1)
```

```python
KS = pd.DataFrame(np.zeros((375,3)))
i = 0
for column in tdata:
    print(i,column)
    KS.loc[i,0] = column
    i = i+1
KS
```

```python
KS.columns = ['field','ks','FDR']
KS
```

```python
import scipy.stats as sps
```

```python
i =0
for column in tdata:
    KS['ks'][i] = sps.ks_2samp(goods[column],bads[column])[0]
    i = i+1
```

```python
KS
```

```python
KS.sort_values(by=['ks'],ascending = False, inplace = True)
KS
```

```python
topRows = int(round(len(tdata)*0.03))
j=0
for column in tdata:
    temp1 = tdata.sort_values(column, ascending = False).head(topRows)
    temp2 = tdata.sort_values(column, ascending = True).head(topRows)
    needed1 = temp1.loc[:,'Fraud']
    needed2 = temp2.loc[:,'Fraud']
    FDR1 = sum(needed1)/numbads
    FDR2 = sum(needed2)/numbads
    FDRate = np.maximum(FDR1,FDR2)
    KS.loc[j, 'FDR'] = FDRate
    j=j+1
    
```

```python
KS.sort_values(by=['ks'], ascending=False, inplace=True)
KS
```
